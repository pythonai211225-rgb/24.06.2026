```python
simple_df = pd.DataFrame([1, 2, 2, 2], ['a', 'b', 'c', 'd'])
simple_df

# which row is a duplication
display(simple_df)
display(simple_df.duplicated())

simple_df[~simple_df.duplicated()]  # get rows without duplication
display(simple_df.drop_duplicates(keep='last')) # get rows without duplicationsame but shorter
display(simple_df.drop_duplicates(keep='first')) # get rows without duplicationsame but shorter
display(simple_df.drop_duplicates(keep=False)) # get rows without duplicationsame but shorter
```

```python
#### missing data - identify and manual remove
df = pd.read_csv('movie_scores.csv')
df.isna()
df[~df['first_name'].isna()]    # 1 - get all rows which have first_name
df[df['first_name'].notnull()]  # 2 - better
df[df['pre_movie_score'].notnull() & df['post_movie_score'].notnull()]
```

```python
#### missing data - identify and manual remove
df = pd.read_csv('movie_scores.csv')
df.dropna()  # <-- remove all rows which have even 1 NaN inside!
df.dropna(thresh=5)  # <-- keep all rows which have at least 5 columns with VALUES inside
```

```python
#### missing data - 
df = pd.read_csv('movie_scores.csv')
df.dropna(subset=['first_name', 'pre_movie_score'])  # remove all rows which have NaN in first_name OR pre_movie_score
```

```python
#### missing data - replace using fillna
df = pd.read_csv('movie_scores.csv')

# 1 manually
df['pre_movie_score'] = df['pre_movie_score'].replace(np.nan, 0)  # manually

# better 
df['pre_movie_score'] = df['pre_movie_score'].fillna(0)  # better

# replace with mean
df['pre_movie_score'] = df['pre_movie_score'].fillna(df['pre_movie_score'].mean())  # better
df
```
